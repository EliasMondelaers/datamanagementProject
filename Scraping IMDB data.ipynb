```python
import time
from imdb import Cinemagoer
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\elias\Desktop\HIRB\Liege\Data Management\final.csv')
```

```python
def get_plot(imdb_id):
    ia = Cinemagoer()
    numeric_id = str(imdb_id).lstrip('tt')
    summary = ""
    storyline = ""
    try:
        movie = ia.get_movie(numeric_id)
        plot = movie.get('plot')
        if plot and len(plot) > 0:
            summary = plot[0]
        if len(plot) > 1:
            storyline = plot[1]
            
    except Exception as e:
        # Optionally, you can print the error so you know it happened:
        # print(f"Error fetching plot for ID {imdb_id}: {e}")
        pass # The 'pass' statement means "do nothing, just continue
    return summary, storyline
```

```python
plot_series = df['imdb_id'].apply(get_plot)

# 2. Convert the Series of tuples into a new DataFrame
plot_data = pd.DataFrame(
    # Use .to_list() to get the list of tuples from the Series
    plot_series.to_list(), 
    # Use the original DataFrame's index to ensure alignment
    index=df.index 
)

# 3. Assign the new columns to the original DataFrame
df['summary'] = plot_data[0]
df['storyline'] = plot_data[1]

# Optional: Drop the temporary DataFrame
plot_data.to_csv(r"C:\Users\elias\Downloads\Data Management\plot_data2.csv")
```
