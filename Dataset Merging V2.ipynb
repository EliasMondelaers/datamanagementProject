```python
df3 = pd.read_csv(r"C:\Users\elias\Desktop\HIRB\Liege\Data Management\archive\movies_metadata.csv")
df3
```

```python
df3['imdb_id'].isnull().sum()
df3['id'] = pd.to_numeric(df3['id'], errors='coerce')

# tt0101704
# tt0358146
# tt0065908
# tt0818098
# tt0176311
# tt1762399
# tt2701582
# tt1325753
# tt0119501
# tt0105298
# tt0493101
# tt13139218
# tt6618142
# tt1147516
# tt0491603
# tt4699464
# tt6051554
```

Impute imdb ID for the missing rows there are only 17 rows to impute

```python
correct_imdb_ids = [
    'tt0101704', 'tt0358146', 'tt0065908', 'tt0818098', 'tt0176311',
    'tt1762399', 'tt2701582', 'tt1325753', 'tt0119501', 'tt0105298',
    'tt0493101', 'tt13139218', 'tt6618142', 'tt1147516', 'tt0491603',
    'tt4699464', 'tt6051554'
]
missing_mask = df3['imdb_id'].isnull()
missing_indices = df3[missing_mask].index
new_ids_series = pd.Series(correct_imdb_ids, index=missing_indices)

df3['imdb_id'].fillna(new_ids_series, inplace=True)
```

```python
merged_df['full_imdbid'] = 'tt0' + merged_df['imdbId'].astype(str)

final_merged_df = pd.merge(
    merged_df,
    df3,
    left_on='full_imdbid',
    right_on='imdb_id',
    how='inner'
)
final_merged_df 
```
